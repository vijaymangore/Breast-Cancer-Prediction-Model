```python
# Split the data into Train and Test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
print("train_size",X_train.shape,y_train.shape)
print("test_size",X_test.shape,y_test.shape)
```

```python
# Data preparation on training data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_transformed=scaler.fit_transform(X_train)
print(X_train_transformed.shape)
```

```python
print("Number of numerical feature",scaler.n_features_in_)
print("mean of each column",scaler.mean_)
print("std of each column",np.sqrt(scaler.var_))
```

```python
#y_train
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y_train_transformed=encoder.fit_transform(y_train)
print(y_train_transformed.shape)
```

```python
y_train_transformed
```

```python
# check the mapping
print(encoder.classes_)
```

```python
y_train.head()
```

```python
# Model building and training
from sklearn.linear_model import LogisticRegression
Log_Reg_model=LogisticRegression()
Log_Reg_model.fit(X_train_transformed,y_train_transformed)
```

```python
#Data preperation on test data
X_test_transformed=scaler.transform(X_test)
print(X_test_transformed.shape)
```

```python
y_test_transformed=encoder.transform(y_test)
print(y_test_transformed.shape)
```

```python
# Predict train and test data
y_train_pred=Log_Reg_model.predict(X_train_transformed)
print(y_train_pred)
y_test_pred=Log_Reg_model.predict(X_test_transformed)
print(y_test_pred)
```

```python
# Evaluation of model performance
from sklearn import metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
print("accuracy_score_train_score:",metrics.accuracy_score(y_train_transformed,y_train_pred))
print("accuracy_score_test_score:",metrics.accuracy_score(y_test_transformed,y_test_pred))
print("precision_score train_score:",precision_score(y_train_transformed,y_train_pred))
print("precision_score test_score:",precision_score(y_test_transformed,y_test_pred))
print("recall_score train_score:",recall_score(y_train_transformed,y_train_pred))
print("recall_score test_score:",recall_score(y_test_transformed,y_test_pred))
print("f1_score train_score:",f1_score(y_train_transformed,y_train_pred))
print("f1_score test_score:",f1_score(y_test_transformed,y_test_pred))
X_train_pred_proba = Log_Reg_model.predict_proba(X_train_transformed)[:, 1]
print("ROC AUC Score train Score:", roc_auc_score(y_train_transformed, X_train_pred_proba))
y_test_pred_proba = Log_Reg_model.predict_proba(X_test_transformed)[:, 1]
print("ROC AUC Score test score:", roc_auc_score(y_test_transformed, y_test_pred_proba))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Example
models = [
    LogisticRegression(),
    RandomForestClassifier(),
    SVC(probability=True),
    KNeighborsClassifier()
]

for model in models:
    model.fit(X_train_transformed, y_train_transformed)
    print(model.__class__.__name__, ":", model.score(X_test_transformed, y_test_transformed))
```
